```python
from databricks_openai import DatabricksOpenAI
client = DatabricksOpenAI()
SUMMARY_MODEL = "databricks-gpt-5-4-mini"
```

```python
def respond(prompt):

    response = client.chat.completions.create(
        model=SUMMARY_MODEL,
        messages=[
            {
                "role": "system",
                "content": (
                    "Respond and greet politely to the user summarizing the request they do"
                )
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.1,
        max_tokens=120
    )

    return response.choices[0].message.content.strip()
```

```python
response = respond("Hello my friend")
print(response)
```
